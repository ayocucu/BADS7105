```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.spatial.distance import cosine
```

```python
df = pd.read_csv('/content/drive/MyDrive/crm/PRD_data.csv')
df = df.drop('ประทับเวลา',1)
df.head(5)
```

```python
df_cs = pd.DataFrame(index=df.columns,columns=df.columns)
for i in range(0,len(df_cs.columns)):
    for j in range(0,len(df_cs.columns)):
        df_cs.iloc[i,j]=1-cosine(df.iloc[:,i],df.iloc[:,j])
df_cs.head(5)
```

```python
df_desc = df.describe().T.sort_values('mean', ascending=False)
df_n = pd.DataFrame({'PRODUCT': list(df_desc.index), 'SCORE': list(df_desc['mean'])})
df_n
```

```python
df_top5 = pd.DataFrame(index=df_cs.columns,columns=range(1,6))
for i in range(0,len(df_top5.columns)):
    df_top5.iloc[i,:5]=df_cs.iloc[0:,i].sort_values(ascending=False)[:5].index
df_top5.head(5)
```
